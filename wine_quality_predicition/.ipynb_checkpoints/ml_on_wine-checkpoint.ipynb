```python
lower_percentile = 10
upper_percentile = 90
```

```python
def clip_outliers(df,col):
    # Calculate percentiles
    lower_threshold = df[col].quantile(lower_percentile / 100)
    upper_threshold = df[col].quantile(upper_percentile / 100)
    df[col] = df[col].clip(lower_threshold, upper_threshold)
```

```python
for col in df.columns:
    clip_outliers(df,col)
```

# Handling imbalanced dataset


#### SMOTE (Synthetic Minority Over-sampling Technique) is an effective oversampling method used to address class imbalance in classification datasets. When dealing with imbalanced data, where the minority class has significantly fewer examples than the majority class, SMOTE helps balance the class distribution by creating synthetic samples for the minority class.

Hereâ€™s how SMOTE works:

It generates new synthetic examples by interpolating between existing minority-class samples.
For each minority-class instance, SMOTE selects its k nearest neighbors (usually k=5).
It creates new synthetic instances by combining the features of the original instance and its neighbors.
SMOTE is particularly useful when:

You want to avoid overfitting caused by random oversampling (simply duplicating minority-class examples).
You need to improve the modelâ€™s performance on the minority class.
Remember that SMOTE is just one of several techniques for handling class imbalance, and it can significantly enhance the performance of machine learning models when applied correctly

```python
from imblearn.over_sampling import SMOTE
```

```python
X = df.drop(['quality'],axis=1)
y= df['quality']
X_res,y_res = SMOTE().fit_resample(X,y)
```

```python
y.value_counts()
```

```python
y_res.value_counts()
```

# Splitting the dataset into the Training set and test set

```python
X_train,X_test,y_train,y_test=train_test_split(X_res,y_res,test_size=0.20,random_state=42)
```

# Feature Scaling

```python
st =StandardScaler()
X_train=st.fit_transform(X_train)
X_test=st.transform(X_test)
```

```python
X_train
```

# Logistic Regression

```python
log = LogisticRegression()
log.fit(X_train,y_train)
```

```python
y_pred1 = log.predict(X_test)
```

```python
accuracy_score(y_test,y_pred1)
```

```python
print(classification_report(y_test, y_pred1))
```

```python
print(confusion_matrix(y_test, y_pred1))
```

# Decision Tree Algorithm

```python
dt =DecisionTreeClassifier()
```

```python
dt.fit(X_train,y_train)
```

```python
y_pred2 = dt.predict(X_test)
```

```python
accuracy_score(y_test,y_pred2)
```

```python
from sklearn.metrics import precision_score
```

```python
precision_score(y_test,y_pred2)
```

```python
from sklearn.metrics import f1_score
```

```python
f1_score(y_test,y_pred2)
```

```python
print(confusion_matrix(y_test, y_pred2))
```

```python
final_data = pd.DataFrame({'Models':['LR','DT'],
             'ACC':[accuracy_score(y_test,y_pred1)*100,
                   accuracy_score(y_test,y_pred2)*100]})
```

```python
final_data
```

```python
sns.barplot(x=final_data['Models'],y=final_data['ACC'])
```
