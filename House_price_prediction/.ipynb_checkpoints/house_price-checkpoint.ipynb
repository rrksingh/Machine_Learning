# Checking for outliers

## histogram and kde plot for all numberical columns

```python
for col in num_col:
    fig, axs=plt.subplots(1,2,figsize=(10,6))
    sns.histplot(data[col], kde=True, ax=axs[0])
    sns.kdeplot(data[col], ax=axs[1])
    print(f'Skewness of {col}: {data[col].skew()}')
```

```python
for col in num_col:
    plt.figure(figsize=(18,4))
    sns.boxplot(data,x=col)
```

```python
df=data.copy()
```

```python
lower_percentile = 10
upper_percentile = 90
```

```python
def clip_outliers(df,col):
    # Calculate percentiles
    lower_threshold = df[col].quantile(lower_percentile / 100)
    upper_threshold = df[col].quantile(upper_percentile / 100)
    df[col] = df[col].clip(lower_threshold, upper_threshold)
```

```python
for col in num_col:
    clip_outliers(df,col)
```

```python
df.shape
```

```python
for col in num_col:
    plt.figure(figsize=(3,4))
    sns.boxplot(x=col,data=df)
```

```python
for col in num_col:
    fig, axs = plt.subplots(1, 2, figsize=(10, 6))  

    sns.histplot(df[col], kde=False, ax=axs[0])  

    sns.kdeplot(df[col], ax=axs[1])
    
    print(f'Skewness of {col}: {df[col].skew()}')
```

```python
for col in num_col:
    plt.figure(figsize=(5,6))
    sns.kdeplot(data[col],color='red')
    sns.kdeplot(df[col])
```

# encoding

```python
cat_col=data.select_dtypes(exclude='number').columns.to_list()
```

```python
cat_col
```

```python
le= LabelEncoder()
for col in cat_col:
    df[col]=le.fit_transform(df[col])
```

```python
df
```

```python
df['price'].dtypes
```

```python
df['price'].unique()
```

```python
X = df.drop(['price'],axis=1)
y= df['price']
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.20,random_state=42)
```

```python
st =StandardScaler()
X_train=st.fit_transform(X_train)
X_test=st.transform(X_test)
```

```python
X_train
```

```python
log = LinearRegression()
log.fit(X_train,y_train)
```

```python
y_pred1 = log.predict(X_test)
```

```python
r2_score(y_test,y_pred1)
```

```python
print(mean_absolute_error(y_test, y_pred1))
```

```python
dt =DecisionTreeRegressor()
```

```python
dt.fit(X_train,y_train)
```

```python
y_pred2 = dt.predict(X_test)
```

```python
r2_score(y_test,y_pred2)
```

```python
mean_absolute_error(y_test,y_pred2)
```

# KNN algorithm

```python
knn=KNeighborsRegressor(n_neighbors=10)
```

```python
knn.fit(X_train,y_train)
```

```python
y_pred3=knn.predict(X_test)
```

```python
r2_score(y_test,y_pred3)
```

```python
final_data = pd.DataFrame({'Models':['LR','DT','Knn'],
             'ACC':[r2_score(y_test,y_pred1)*100,
                   r2_score(y_test,y_pred2)*100,
                   r2_score(y_test,y_pred3)*100]})
```

```python
final_data
```

```python
sns.barplot(x=final_data['Models'],y=final_data['ACC'])
```
