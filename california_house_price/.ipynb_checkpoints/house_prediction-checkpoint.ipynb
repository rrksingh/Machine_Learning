# slove outliers 

```python
lower_percentile = 10
upper_percentile = 90
```

```python
def clip_outliers(df,col):
    # Calculate percentiles
    lower_threshold = df[col].quantile(lower_percentile / 100)
    upper_threshold = df[col].quantile(upper_percentile / 100)
    df[col] = df[col].clip(lower_threshold, upper_threshold)
```

```python
for col in cols:
    clip_outliers(df,col)
```

```python
for col in num_col:
    plt.figure(figsize=(3,4))
    sns.boxplot(x=col,data=df)
```

```python
X = df.drop(['median_house_value'],axis=1)
y= df['median_house_value']
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.20,random_state=42)
```

```python
st =StandardScaler()
X_train=st.fit_transform(X_train)
X_test=st.transform(X_test)
```

```python
X_train
```

```python
log = LinearRegression()
log.fit(X_train,y_train)
```

```python
y_pred1 = log.predict(X_test)
```

```python
r2_score(y_test,y_pred1)
```

```python
print(mean_absolute_error(y_test, y_pred1))
```

```python
dt =DecisionTreeRegressor()
```

```python
dt.fit(X_train,y_train)
```

```python
y_pred2 = dt.predict(X_test)
```

```python
r2_score(y_test,y_pred2)
```

```python
mean_absolute_error(y_test,y_pred2)
```

```python
knn=KNeighborsRegressor()
```

```python
param_grid = {
    'n_neighbors': [1,2,3, 5, 7, 9, 11],
    'weights': ['uniform', 'distance'],
    'algorithm': ['auto', 'ball_tree', 'kd_tree', 'brute'],
    'p': [1, 2]
}
```

```python
grid_search = GridSearchCV(estimator=knn, param_grid=param_grid, cv=5, n_jobs=-1, verbose=2)
grid_search.fit(X_train, y_train)
```

```python
knn.fit(X_train,y_train)
```

```python
y_pred3=grid_search.predict(X_test)
```

```python
r2_score(y_test,y_pred3)
```

```python
final_data = pd.DataFrame({'Models':['LR','DT','Knn'],
             'ACC':[r2_score(y_test,y_pred1)*100,
                   r2_score(y_test,y_pred2)*100,
                   r2_score(y_test,y_pred3)*100]})
```

```python
final_data
```

```python
sns.barplot(x=final_data['Models'],y=final_data['ACC'])
```
